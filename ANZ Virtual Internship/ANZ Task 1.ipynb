```python
M_toexp=0
F_toexp=0
for ind in df.index:
    if df['gender'][ind]=='M':
        M_toexp=M_toexp+df['balance'][ind]
    else:
        F_toexp=F_toexp+df['balance'][ind]
exp_list=[M_toexp,F_toexp]
label_list=['Male Expenditure','Female Expenditure']
```
