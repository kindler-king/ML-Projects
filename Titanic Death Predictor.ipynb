```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('titanic.csv')
data
```

```python
data.drop(columns=['PassengerId','Name','SibSp','Parch','Age','Ticket','Cabin','Embarked'],inplace=True)
```

```python
data.columns
```

```python
data['Sex'].replace({'male':1,'female':0},inplace=True)
```

```python
data['Sex'].value_counts()
```

```python
x=data.iloc[:,1:].values
```

```python
x.shape
```

```python
y=data.iloc[:,0].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=7)
```

```python

from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(x_train,y_train)
```

```python
x_train.shape
```

```python
plt.imshow(x_test[100].reshape(1,3))
classifier.predict(x_test[100].reshape(1,3))
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
